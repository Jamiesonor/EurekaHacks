```python
import matplotlib.pyplot as plt
import json

def plot_pie_counter(title, counter):
    labels = list(counter.keys())
    sizes = counter.values()
    plt.figure(figsize=(7, 5))
    plt.title(title)
    patches, texts = plt.pie(sizes, shadow=True, startangle=90)
    plt.legend(patches, labels, loc='best')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
    print(title, json.dumps(counter, sort_keys=False, indent=4))

plot_pie_counter(
    'Labels',
    installerErrorCounts,
)
```
